````python
json_data = re.sub(r'[\x00-\x1F\x7F]', '', json_data)

````

````python


data = json.loads(json_data)
#print(data) 

message_map = data.get('data').get('list') 
comment_list = []

for item in message_map:
    message = item.get('message')
    if message:  # optional check if message is not None or empty
        comment_list.append(message)

print(len(comment_list))
````

````python

aci_url = "http://"+aci_fqdn+":5000/text/analytics/v3.1/sentiment"


headers = {
    "Content-Type": "application/json"
}

data = {
    "documents": [
        {
            "id": idx + 1,
            "text": comment
        }
        for idx, comment in enumerate(comment_list)
    ]
}
# print(data)

response = requests.post(aci_url, headers=headers, data=json.dumps(data))
print(response.status_code)
#print(response.json())  # or response.text if not JSON
sentiments = response.json()
````

````python
for doc in sentiments['documents']:
    doc.pop('sentences', None)
    doc.pop('confidenceScores', None)
    doc.pop('warnings', None)

# Result
#print(sentiments)

# Step 1: Create a lookup dictionary for sentiments
sentiment_lookup = {int(doc['id']): doc['sentiment'] for doc in sentiments['documents']}

# Step 2: Construct the result list
result = []
for doc in data['documents']:
    doc_id = doc['id']
    if doc_id in sentiment_lookup:
        result.append({
            'id': doc_id,
            'text': doc['text'],
            'sentiment': sentiment_lookup[doc_id]
        })

# Optional: wrap in a dictionary
result_dict = {'documents': result}

# Output
#print(result_dict)

import pandas as pd 
df = pd.DataFrame(result_dict['documents'])
pd.set_option('display.max_colwidth', None)
df
````
